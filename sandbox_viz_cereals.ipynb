```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cereal_df = pd.read_csv("cereal.csv")
```

```python
cereal_df
```

```python
cereal_df["fiber"].hist(bins=14)
plt.show()
```

```python
# Needed something categorical, used the 'mfr' column (Manufacturer) and got calories per mfr.
man_cereal_df = cereal_df.groupby("mfr")["calories"].mean()
print(man_cereal_df)
```

```python
# can use kind="line" if using time data
man_cereal_df.plot(kind="bar", rot=0)
plt.title("Avg. Calories per Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Calories")
plt.show()
```

```python
# scatter
cereal_df.plot(x="sugars", y="fat", kind="scatter")
plt.show()
```

```python
avocados = pd.read_csv("avocado.csv")
```

```python
avocados.head()
```

```python
pd.unique(avocados["type"])
```

```python
pd.unique(avocados["region"])
```

```python
# I can calculate the totals in each column, I don't need figures telling me.
avocados.drop(avocados[avocados.region == "TotalUS"].index, inplace=True)
```

```python
pd.unique(avocados["region"])
```

```python
mean_av_region_total = avocados.groupby("region")["Total Bags"].mean()
```

```python
plt.figure(figsize=(18,6))
mean_av_region_total.plot(kind="bar")
plt.show()
```

```python
av_sold_by_date_total = avocados.groupby("Date")["Total Bags"].sum()
av_sold_by_date_small = avocados.groupby("Date")["Small Bags"].sum()
av_sold_by_date_large = avocados.groupby("Date")["Large Bags"].sum()
av_sold_by_date_xlarge = avocados.groupby("Date")["XLarge Bags"].sum()
```

```python
#no 'type' as 'plot' defaults to line type.
# comparing all types of bag sales over time.
plt.figure(figsize=(19,8))
plt.tight_layout()
plt.style.use('seaborn')
av_sold_by_date_small.plot()
av_sold_by_date_large.plot()
av_sold_by_date_xlarge.plot()
plt.legend()
plt.title("Avocado Bag Size Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Bags Sold (1,000's)")
plt.grid()
plt.show()
```
